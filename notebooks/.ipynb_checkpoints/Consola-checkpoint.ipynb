# Consola

#### En este cuaderno vamos a preparar los archivos para ver que el código se ejecuta por consola correctamente.

````python
import pandas as pd
````

````python
train = pd.read_csv('../data/train.csv')
test = pd.read_csv('../data/test.csv')
````

````python
display(train)
````

````python
display(test)
````

#### Tomemos una pequeña muestra del conjunto de datos para ejecutar por consola.

````python
train_small = train[:1000]
test_small = test[:200]
````

````python
train_small.to_csv('../data/train-small.csv', index=False)
test_small.to_csv('../data/test-small.csv', index=False)
````

#### Ejecutemos por consola usando ambos métodos y guardemos el resultado. Entonces, abrimos la consola en la carpeta principal del trabajo y ejecutamos:

```
build/tp2 -m 0 -i data/train-small.csv -q data/test-small.csv -i data/result-0.csv
build/tp2 -m 1 -i data/train-small.csv -q data/test-small.csv -i data/result-1.csv
```
#### Cuando termine de ejecutar ambos métodos vamos a tener dos archivos csv, uno llamado `result-0.csv` con el resultado del método que solo usa kNN (con k = 10) y otro llamado `result-1.csv` con el resultado del método que usa PCA (con \\( \alpha \\) = 30) y kNN (con k = 10). Para ver todas las opciones que hay para ejecutar por consola podemos correr `cmake-build-debug/tp2 --help`.
